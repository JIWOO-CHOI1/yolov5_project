```python
import cv2
import torch
from PIL import Image
import os

# YOLOv5 모델 로드
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/workspace/LBW-Intern/웅기/yolov5/runs/train/woong_yolov5m_results2/weights/best.pt')

# 입력 동영상 파일이 있는 폴더 경로
input_folder = '/workspace/LBW-Intern/지우/지우_test/'

# 결과 동영상 파일이 저장될 폴더 경로
output_folder = '/workspace/LBW-Intern/웅기/결과파일/'

# 동영상 파일 목록 가져오기
video_files = [f for f in os.listdir(input_folder) if f.endswith('.mp4')]

# 반복문을 사용하여 각 동영상 파일에 대해 작업 수행
for video_file in video_files:
    input_path = os.path.join(input_folder, video_file)

    # 동영상 불러오기
    video = cv2.VideoCapture(input_path)

    # 동영상 해상도 및 FPS 정보 가져오기
    frame_width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = video.get(cv2.CAP_PROP_FPS)

    # 동영상 저장을 위한 VideoWriter 설정
    output_path = os.path.join(output_folder, f'result_{video_file}')
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    output_video = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    while True:
        ret, frame = video.read()

        if not ret:
            break

        # 프레임을 PIL 이미지로 변환
        pil_image = Image.fromarray(frame)

        # YOLOv5를 사용하여 객체 탐지 수행
        results = model(pil_image)

        # 탐지 결과를 프레임에 표시
        output_frame = results.render()[0]  # 결과 이미지로 변환
        output_frame = cv2.cvtColor(output_frame, cv2.COLOR_RGB2BGR)  # OpenCV의 BGR 형식으로 변환
        output_video.write(output_frame)

    # 동영상 처리 완료 후 리소스 해제
    video.release()
    output_video.release()
```
