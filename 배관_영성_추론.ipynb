```python
import cv2
import torch
from PIL import Image
import os

# YOLOv5 ëª¨ë¸ ë¡œë“œ
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/workspace/LBW-Intern/ì›…ê¸°/yolov5/runs/train/woong_yolov5m_results2/weights/best.pt')

# ì…ë ¥ ë™ì˜ìƒ íŒŒì¼ì´ ìˆëŠ” í´ë” ê²½ë¡œ
input_folder = '/workspace/LBW-Intern/ì§€ìš°/ì§€ìš°_test/'

# ê²°ê³¼ ë™ì˜ìƒ íŒŒì¼ì´ ì €ì¥ë  í´ë” ê²½ë¡œ
output_folder = '/workspace/LBW-Intern/ì›…ê¸°/ê²°ê³¼íŒŒì¼/'

# ë™ì˜ìƒ íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
video_files = [f for f in os.listdir(input_folder) if f.endswith('.mp4')]

# ë°˜ë³µë¬¸ì„ ì‚¬ìš©í•˜ì—¬ ê° ë™ì˜ìƒ íŒŒì¼ì— ëŒ€í•´ ì‘ì—… ìˆ˜í–‰
for video_file in video_files:
    input_path = os.path.join(input_folder, video_file)

    # ë™ì˜ìƒ ë¶ˆëŸ¬ì˜¤ê¸°
    video = cv2.VideoCapture(input_path)

    # ë™ì˜ìƒ í•´ìƒë„ ë° FPS ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    frame_width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = video.get(cv2.CAP_PROP_FPS)

    # ë™ì˜ìƒ ì €ì¥ì„ ìœ„í•œ VideoWriter ì„¤ì •
    output_path = os.path.join(output_folder, f'result_{video_file}')
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    output_video = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    while True:
        ret, frame = video.read()

        if not ret:
            break

        # í”„ë ˆì„ì„ PIL ì´ë¯¸ì§€ë¡œ ë³€í™˜
        pil_image = Image.fromarray(frame)

        # YOLOv5ë¥¼ ì‚¬ìš©í•˜ì—¬ ê°ì²´ íƒì§€ ìˆ˜í–‰
        results = model(pil_image)

        # íƒì§€ ê²°ê³¼ë¥¼ í”„ë ˆì„ì— í‘œì‹œ
        output_frame = results.render()[0]  # ê²°ê³¼ ì´ë¯¸ì§€ë¡œ ë³€í™˜
        output_frame = cv2.cvtColor(output_frame, cv2.COLOR_RGB2BGR)  # OpenCVì˜ BGR í˜•ì‹ìœ¼ë¡œ ë³€í™˜
        output_video.write(output_frame)

    # ë™ì˜ìƒ ì²˜ë¦¬ ì™„ë£Œ í›„ ë¦¬ì†ŒìŠ¤ í•´ì œ
    video.release()
    output_video.release()
```
