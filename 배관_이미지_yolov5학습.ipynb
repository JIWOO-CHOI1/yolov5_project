```python
from glob import glob

img_list = glob('/workspace/LBW-Intern/웅기/images/*.jpg')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
with open('/workspace/LBW-Intern/웅기/train.txt', 'w') as f:
    f.write('\n'.join(train_img_list) + '\n')
with open('/workspace/LBW-Intern/웅기/val.txt', 'w') as f:
    f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml
```

```python
with open('/workspace/LBW-Intern/웅기/images/data.yaml', 'r')  as f:
    data = yaml.full_load(f)

print(data)

data['train'] = '/workspace/LBW-Intern/웅기/train.txt'
data['val'] = '/workspace/LBW-Intern/웅기/val.txt'

with open('/workspace/LBW-Intern/웅기/images/data.yaml', 'w') as f:
    yaml.dump(data, f)

print(data)
```
